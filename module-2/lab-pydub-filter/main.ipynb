```python
from pydub import AudioSegment
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
song = AudioSegment.from_mp3('./input/input.mp3')
samples = song.get_array_of_samples()
```

```python
canales = song.channels
frame_rate = song.frame_rate
duracion = song.duration_seconds
sample_width = song.sample_width
```

```python
plt.plot(samples)
```

```python
b, a = signal.butter(8, 0.05)
y = signal.filtfilt(b, a, samples)
```

```python
plt.plot(y)
```

```python
audio_segment = AudioSegment(
    y.tobytes(), 
    frame_rate=frame_rate,
    sample_width=sample_width,
    channels=canales
)
```

```python
audio_segment.export('./output/output.mp3',format='mp3')
```
